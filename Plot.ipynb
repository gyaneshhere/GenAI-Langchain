```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the tickers
stocks = ['AMZN', 'MSFT', 'NVDA', '^GSPC']  # ^GSPC is the S&P 500

# Fetch historical data for the last 5 years
end_date = pd.Timestamp.now()
start_date = end_date - pd.DateOffset(years=5)

# Download stock data
stock_data = yf.download(stocks, start=start_date, end=end_date)

# Calculate percentage gain for each stock
percentage_gains = ((stock_data['Adj Close'].iloc[-1] - stock_data['Adj Close'].iloc[0]) / stock_data['Adj Close'].iloc[0]) * 100

# Plot
plt.figure(figsize=(10, 6), facecolor='white')
percentage_gains.plot(kind='bar')
plt.title('Percentage Gain of Each Stock Over Last 5 Years')
plt.ylabel('Percentage Gain')
plt.xlabel('Stock')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

print('Percentage gains calculated and plotted.')
```
